```python
#Importing libraries
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import random
import bisect 
import collections
```

```python
#Website and 
url ="https://oldschool.runescape.wiki/w/Zulrah"
response = requests.get(url,headers={'User-Agent':'Mozilla/5.0'})
page_content = response.content
soup = BeautifulSoup(page_content,'lxml')
print(soup)
```

```python
my_tables = soup.find_all('table',attrs= {"class":"wikitable sortable filterable item-drops autosort=4,a"})
print(my_tables)
```

```python
#Getting the data from table
loot_tables = {}
table_index =0

for table in my_tables:
    table_r = table.find_all('tr')
    
    table_h = table.find_all('th')
    
    cols=[]
    for item in table_h:
        if len(item.text)>0:
            cols.append(item.text)
        
    data = []
    for row in table_r:
        table_data = row.find_all('td')
        
        row_data = [item.text for item in table_data if item.text != '']
        
        if len(row_data)> 0:
            data.append(row_data)
            
    output_frame= pd.DataFrame(data,columns = cols)        
    print(output_frame)    
    
    loot_tables.setdefault(table_index, output_frame)
    table_index += 1
    
    #print(data)
    #print(cols)
```
